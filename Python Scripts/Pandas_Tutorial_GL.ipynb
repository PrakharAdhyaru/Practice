```python
df7 = pd.DataFrame({'customer':[101,102,103,104],
                    'category':['cat2','cat2','cat1','cat3'] ,
                    'important':['yes','no','yes','yes'],
                    'sales':[123,52,214,663]},index=[0,1,2,3])
df8 = pd.DataFrame({'customer':[101,103,104,105],
                    'color':['yellow','green','green','blue'] ,
                    'distance':[12,9,44,21],
                    'sales':[123,214,663,331]},index=[4,5,6,7])
```

```python
pd.concat([df7,df8],axis=0)
```

```python
pd.concat([df7,df8],axis=0,sort=True)
```

```python
pd.concat([df7,df8],axis=1)
```

```python
pd.concat([df7,df8],axis=1,sort=True)
```

```python
# Merge 
# how{left, right, outer, inner, cross}, default inner
# on = 'coloumn name' - (axis=1)
# pandas.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, 
# sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)

pd.merge(df7,df8,how='outer',on='customer')  # outer merge is an union of 'on'
#pd.merge(df7,df8,how='outer',on='customer', suffixes=('_df7', '_df8'))
```

```python
pd.merge(df7,df8,how='inner',on='customer')  # inner merge is an intersection of 'on'
```

```python
pd.merge(df7,df8,how='left',on='customer')  # left merge is just the first an , but all columns 
```

```python
pd.merge(df7,df8,how='right',on='customer')  # right merge is just the second an , but all columns 
```
